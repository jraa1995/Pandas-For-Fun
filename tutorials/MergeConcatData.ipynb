## Merging & Concatenating Data

```python
import pandas as pd
bios = pd.read_csv('../data/bios.csv')
```

```python
# one thing to note is that some athletes may be born in a different country than where they are competing
bios.head()
```

```python
# lets say we want to compare where they were born and where they are competing / were competing
nocs = pd.read_csv('../data/noc_regions.csv')
nocs.head()
```

```python
# we can merge this data with BIOS. 
# you could also change the suffix so instead of it assigned an auto suffix and show NOC_x, NOC_y
# we can do
# bios_update = pd.merge(bios, nocs, left_on='born_country', right_on='NOC', how='left', suffixes=["bios", "nocdf"])
bios_update = pd.merge(bios, nocs, left_on='born_country', right_on='NOC', how='left')
```

```python
bios_update.rename(columns={'region': 'born_country_full'}, inplace=True)
bios_update
```

```python
# lets compare
bios_update[bios_update['NOC_x'] != bios_update['born_country_full']][['name', 'NOC_x', 'born_country_full']]
```

```python
# what if we just want to take the athletes from USA? GBR?
usa = bios[bios['born_country']=='USA'].copy()
gbr = bios[bios['born_country']=='GBR'].copy()
```

```python
# now, lets make a new DF that only contains usa and gbr
new_df = pd.concat([usa, gbr])
new_df
```
